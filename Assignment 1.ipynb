# Data cleaning

```python
df = df.drop(columns=['Reference'])
df = df.drop(columns=['Number and type of monitoring stations'])
df = df.drop(columns=['Version of the database'])
df = df.drop(columns=['Status'])


# Impute missing values in numerical columns using the mean
numerical_cols = ['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)', 'PM25 temporal coverage (%)', 'PM10 temporal coverage (%)', 'NO2 temporal coverage (%)']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify changes
display(df.info())
display(df.describe())
```

```python
# Create box plots to identify outliers in numerical columns
numerical_cols = ['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)', 'PM25 temporal coverage (%)', 'PM10 temporal coverage (%)', 'NO2 temporal coverage (%)']
plt.figure(figsize=(15, 10))  # Adjust figure size for better visualization

for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    plt.boxplot(df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
#Filtering China's data
CHN_df = df[df['ISO3'] == 'CHN']
CHN_df = CHN_df.drop(columns=['PM10 (μg/m3)'])
CHN_df = CHN_df.drop(columns=['NO2 (μg/m3)'])
CHN_df = CHN_df.drop(columns=['PM10 temporal coverage (%)'])
CHN_df = CHN_df.drop(columns=['NO2 temporal coverage (%)'])

CHN_df.describe()
```

```python
# Create box plots to identify outliers in numerical columns
numerical_cols = ['PM2.5 (μg/m3)', 'PM25 temporal coverage (%)']
plt.figure(figsize=(15, 10))  # Adjust figure size for better visualization

for i, col in enumerate(numerical_cols):
    plt.subplot(1, 2, i + 1)
    plt.boxplot(CHN_df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Calculate the 50th percentile (Q2)
Q2 = CHN_df['PM25 temporal coverage (%)'].quantile(0.50)

# Filter the DataFrame to keep only values above Q2
CHN_df = CHN_df[CHN_df['PM25 temporal coverage (%)'] > Q2]

# Recreate the box plots
numerical_cols = ['PM2.5 (μg/m3)', 'PM25 temporal coverage (%)']
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(1, 2, i + 1)
    plt.boxplot(CHN_df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

# Global Trend Analysis

```python
import seaborn as sns

# Global Trend Analysis

# Trend of air pollution indicators
plt.figure(figsize=(15, 10))
for i, col in enumerate(['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)']):
    plt.subplot(2, 2, i + 1)
    df.groupby('Measurement Year')[col].mean().plot(kind='line')
    plt.title(f'Trend of {col} over time')
    # Limit the x-axis to the range 2010-2019
    plt.xlim(2010, 2019)
plt.tight_layout()

plt.show()
```

```python
# Trend of air pollution indicators
plt.figure(figsize=(15, 10))
for i, col in enumerate(['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)']):
    plt.subplot(3, 1, i + 1)
    # Apply 3-year rolling mean for smoothing
    df.groupby('Measurement Year')[col].mean().rolling(window=3).mean().plot(kind='line', linestyle='--', color='blue', linewidth=2)
    plt.title(f'Trend of {col} over time')
    plt.xlim(2010, 2019)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

# Select the columns representing temporal coverage percentages
temporal_cols = ['PM25 temporal coverage (%)', 'PM10 temporal coverage (%)', 'NO2 temporal coverage (%)']

for col in temporal_cols:
    df.groupby('Measurement Year')[col].mean().plot(kind='line', label=col)

plt.title('Temporal Percentage Trend of Air Pollution Indicators Over Time')
plt.xlabel('Measurement Year')
plt.ylabel('Temporal Percentage (%)')
plt.legend()
plt.grid(True)

# Limit the x-axis to the range 2010-2019
plt.xlim(2010, 2019)

plt.show()
```

```python
temporal_cols = ['PM25 temporal coverage (%)', 'PM10 temporal coverage (%)', 'NO2 temporal coverage (%)']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, col in enumerate(temporal_cols):
    axes[i].scatter(df['Measurement Year'], df[col])
    correlation = np.corrcoef(df['Measurement Year'], df[col])[0, 1]
    axes[i].text(0.05, 0.95, f'Correlation: {correlation:.2f}', transform=axes[i].transAxes,
                 fontsize=10, verticalalignment='top')
    axes[i].set_title(f'{col} vs. Measurement Year')
    axes[i].set_xlabel('Measurement Year')
    axes[i].set_ylabel(col)
    axes[i].grid(True)
    axes[i].set_xlim(2010, 2019)  # Limit x-axis

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Group data by year and calculate the mean PM2.5 for each year
pm25_by_year = CHN_df.groupby('Measurement Year')['PM2.5 (μg/m3)'].mean()

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(pm25_by_year.index, pm25_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average PM2.5 (μg/m3)')
plt.title('Average PM2.5 in China over Time')
plt.grid(True)
plt.show()
```

```python
# Histogram for China in 2019
plt.figure(figsize=(12, 6))
plt.hist(CHN_df['PM2.5 (μg/m3)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of PM2.5 in China (2019)')
plt.xlabel('PM2.5 (μg/m3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Filter data for the year 2019
CHN_df_2019 = CHN_df[CHN_df['Measurement Year'] == 2019]

# Group data by city and calculate the mean PM2.5 for each city
pm25_by_city = CHN_df_2019.groupby('City or Locality')['PM2.5 (μg/m3)'].mean()


# Histogram for PM2.5 in each city
plt.figure(figsize=(12, 6))
plt.hist(pm25_by_city, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of PM2.5 in Chinese Cities (2019)')
plt.xlabel('PM2.5 (μg/m3)')
plt.ylabel('Frequency')
plt.show()


# Donut chart for PM2.5 contribution of each city
# Calculate the total PM2.5 for all cities in 2019
total_pm25 = pm25_by_city.sum()
# Calculate percentage contribution of each city
city_percentages = (pm25_by_city / total_pm25) * 100

# Create the donut chart
plt.figure(figsize=(8, 8))
plt.pie(city_percentages, labels=city_percentages.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
plt.title('Percentage Contribution of PM2.5 by City (2019)')

# Add a circle in the center to create the donut shape
centre_circle = plt.Circle((0, 0), 0.60, fc='white')  # Adjust radius as needed
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')

plt.show()
```

```python
# prompt: draw donut chart for top 5 cities in 2019, then compare with others

import matplotlib.pyplot as plt

# Assuming pm25_by_city is already calculated as in the original code
# Filter for top 5 cities
top_5_cities = pm25_by_city.nlargest(5)
other_cities = pm25_by_city.drop(top_5_cities.index)

# Calculate total PM2.5 for top 5 and others
total_pm25_top5 = top_5_cities.sum()
total_pm25_others = other_cities.sum()

# Create the donut chart
plt.figure(figsize=(8, 8))

# Data for the donut chart
sizes = [total_pm25_top5, total_pm25_others]
labels = ['Top 5 Cities', 'Other Cities']
colors = ['skyblue', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
plt.title('PM2.5 Contribution: Top 5 vs. Others (2019)')

# Add a circle in the center to create the donut shape
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')

plt.show()

# Comparison bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_5_cities.index, top_5_cities.values, label='Top 5 Cities', color='skyblue')
plt.axhline(y=total_pm25_others / len(other_cities), color='lightcoral', linestyle='--', label='Average of Other Cities')
plt.xlabel('City')
plt.ylabel('Average PM2.5 (μg/m3)')
plt.title('Top 5 Cities vs. Average of Other Cities (2019)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()
plt.tight_layout()
plt.show()
```

# Global Perspectives

```python
import matplotlib.cm as cm

indicators = ['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)']

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

# Get unique regions and assign colors
regions = df['WHO Region'].unique()
num_regions = len(regions)
colors = cm.get_cmap('viridis', num_regions).colors

# Store handles and labels for legend
handles = []
labels = []

for i, indicator in enumerate(indicators):
    for j, region in enumerate(regions):
        region_data = df[df['WHO Region'] == region].groupby('Measurement Year')[indicator].mean()

        # Check if region_data is empty before creating the bar
        if not region_data.empty:
            # Create bar and store handle for legend
            bar = axes[i].bar(region_data.index, region_data.values, label=region, alpha=0.7, color=colors[j])

            # Add handle and label to lists for legend
            if region not in labels:  # Only add unique legend entries
                handles.append(bar[0])  # Get the first bar object for the legend
                labels.append(region)

    axes[i].set_xlabel('Year')
    axes[i].set_xticks(range(2010, 2020))
    axes[i].set_ylabel(indicator)
    axes[i].set_title(f'Regional {indicator} over Time')

    # Create the legend using the collected handles and labels
    axes[i].legend(handles, labels)

    axes[i].grid(True)
    axes[i].set_xlim(2009, 2020)
    plt.margins(x=0.05)

plt.tight_layout()
plt.show()
```

```python
indicators = ['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns for pie charts

for i, indicator in enumerate(indicators):
    # Calculate total for each country for the indicator
    country_totals = df.groupby('ISO3')[indicator].sum().sort_values(ascending=False)

    # Get top 5 countries
    top_5_countries = country_totals[:5]

    # Calculate 'Others' total
    others_total = country_totals[5:].sum()

    # Combine top 5 and 'Others'
    pie_data = pd.concat([top_5_countries, pd.Series([others_total], index=['Others'])])
    # Create pie chart
    axes[i].pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Top 5 Contributors to {indicator}')

plt.tight_layout()
plt.show()
```
